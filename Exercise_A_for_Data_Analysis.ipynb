Selection Based on Conditionals

Let's see which stations have the most and fewest rides.


```python
# use the min() and max() modules
print(ridership['rides'].min())

print('------------------------------------')

print(ridership['rides'].max())
```

```python
ridership.columns
```

```python
filter=ridership['rides']==ridership['rides'].max()
max_rides_rows = ridership[filter]
print(max_rides_rows)
```

Create a Time Index

So we have all rides from January 1, 2001 to December 31, 2019. Let's do some statistics that group ride statistics by date. First we need to tell Pandas that the date column is in fact a date. So, we convert the column to a proper 'DateTime' type, and then set the index to this column.

This step takes a little bit of time!

The following picture show the desired output

image.png

```python
# hint, use to_datetime() and set_index()  modules
ridership['date'] = pd.to_datetime(ridership['date'])
```

```python
ridership.head(10)
```

Sort the Dataframe by Date

```python
# hint, use to_datetime() and set_index()  modules
# your code here
#pass
ridership=ridership.set_index('date') #Establece la columna 'fecha' como el ndice del DataFrame
```

```python
ridership.head(10)
```

```python
# hint, sort_values(by='date')
ridership=ridership.sort_values(by='date')
```

```python
ridership.head(10)
```

Sanity Checking

Looks good! Now let's have a quick look at data for specific stations:
'Garfield-Dan Ryan' and 'Garfield-South Elevated'


```python
# Assuming 'ridership' is your DataFrame
filter_garfield_dan_ryan = ridership['stationname'] == 'Garfield-Dan Ryan'
# Applying the filter to get rows for 'Garfield-Dan Ryan'
Garfield_Dan_Ryan = ridership[filter_garfield_dan_ryan]
print(Garfield_Dan_Ryan)
```

```python
# Filtering for 'Garfield-South Elevated'
filter_garfield_south_elevated = ridership['stationname'] == 'Garfield-South Elevated'
# Applying the filter to get rows for 'Garfield-South Elevated'
Garfield_South_Elevated = ridership[filter_garfield_south_elevated]
print("\nGarfield-South Elevated Station:")
print(Garfield_South_Elevated)
```

Visualizing Timeseries Data

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'Garfield_Dan_Ryan' is your DataFrame
sns.set(rc={'figure.figsize': (10, 7)})

# Plotting the 'rides' column
plt.plot(Garfield_Dan_Ryan['rides'], linewidth=0.5)

# Adding labels and title
plt.xlabel('Date')  # Assuming the x-axis represents dates
plt.ylabel('Number of Rides')
plt.title('Rides over Time for Garfield-Dan Ryan Station')

# Display the plot
plt.show()
```

```python
sns.set(rc={'figure.figsize': (10, 7)})
plt.plot(Garfield_South_Elevated['rides'], linewidth=0.5)
plt.xlabel('Date')  # Assuming the x-axis represents dates
plt.ylabel('Number of Rides')
plt.title('Rides over Time for Garfield-South Elevated Station')
plt.show()
```
